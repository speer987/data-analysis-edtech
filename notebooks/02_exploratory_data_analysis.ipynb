```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis (EDA)

### Loading the Merged Datasets

```python
df = pd.read_csv('../data/cleaned/cleaned_data.csv')
```

Let's check to see the data types before proceeding with the exploration aspect, since we need to check if it got imported properly.

```python
df.info()
```

Since the Postal Code and Electric Range in Miles column turned into floats again (because of NaN values), let's convert the float values to Int64 (a nullable integer) again:

```python
df['Postal Code'] = df['Postal Code'].astype('Int64')
df['Electric Range (in Miles)'] = df['Electric Range (in Miles)'].astype('Int64')
df.head()
```

### Exploratory Data Analysis and Visual Analyses

We can get the summary statistics about this dataset by doing the following:

```python
df.describe()
```

I also want to check how many null values there are:

```python
df.isnull().sum()
```

Seeing the above data, I realized that for the "Electric Range" and "Clean Alternative Fuel (CAFV) Eligible" columns, 177937 rows of the data are null out of 279780 rows (64%). 

As such, we will have to deal with the analysis done on those rows seperately, so that any queries that involve those columns are only done on a subset of the dataset, which is the 36% of the data that doesn't have null values for these columns.

```python
# The following line will give you a vector of true/false values, which we can use to make a new dataframe of only those that say that there is data, hence the .notnull().
df['Electric Range (in Miles)'].notnull()
# We can define a new dataframe using the following:
df_elec = df[df['Electric Range (in Miles)'].notnull()]
df_elec.head()
```

We can use the df_elec that we made when we have to analyze data that has values for the Electric Range column. This is useful in case we need to analyze only cars that have Electric Ranges.

We can also take a look at the unique values of the dataset using the following:

```python
df.nunique()
```

After seeing that there are 52 unique states included in this dataset, I realized that I joined the EV dataset with an unsuitable gas dataset the first time. This is because the gas dataset only includes information about WA gas prices, but the EV dataset contains multiple states in it. 

As I was trying to find a new gas dataset to replace the gas dataset with, I found that when I merged the gas dataset and the EV dataset in the data cleaning file, I didn't account for differences in dataset granularity.

Merging each row in the EV dataset with a gas price joined on the year column would be inappropriate because there are only 18 gas prices, but each row in the EV dataset is a seperate car. 

This would cause a misrepresentation of gas price data because the gas price would be the same across multiple rows depending on the year and has no connection with each car. This would skew the data.

As such, I made the decision to drop the gas prices column, although it could be accounted for using aggregate values if this project was extended.

```python
df = df.drop(columns='WA Retail Gas Prices (Dollars per Gallon)')
df.to_csv("../data/cleaned/data_wo_fuel.csv", index=False)
```

In order to check the cars that have the highest electric range in miles, I did the following:

```python
df.sort_values(by='Electric Range (in Miles)', ascending=False).head()
```

We can see that most of the cars in the top few are Teslas and sometimes they are also of different models.

I wanted to also check if there is a correlation between the electric range and the model year as well, so we could do the following in order to see how highly correlated the numerical data is to encourage use to investigate certain features of the dataframe further.

```python
df.corr(numeric_only=True)
```

All the correlations are extremely low, so this shows us that there isn't much of a relationship between Model Year and postal code for instance. This goes for all the numerical values in this dataset it seems. In order to see what these relationships look like graphically, I decided to use a heatmap from Seaborn. 

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.rcParams['figure.figsize'] = (5, 5)
plt.show()
```

As expected, the correlations are shown in the negative, hence most of the squares being nearly purple or black. 

I was also curious about the most popular brands of cars in each state, so I decided to group by the state and make of the car, and counted how many electric cars there are per state with the following:

```python
df.groupby(['State', 'Make'])['Make'].count().sort_values(ascending=False)
```

This confirms out observation earlier, that most of the cars in this dataset are centered around Washington State, but it also confirms our discussion Teslas being the leader when it comes to electric vehicles in terms of how many cars there are in the EV population dataset. Chevrolets seem to be the second most registered car, with Nissan and others following.

I also thought it would be interesting to plot how the adoption of BEVs and PHEVs changes over the years. I decided to first group the cars by the year, and the 'Electric Vehicle Type' column before counting each type of EV. Finally, I unstacked the table, as this would allow me to see the number of PHEVs and BEVs side by side per year in a dataframe.

```python
df_ev_type = df.groupby(['Model Year', 'Electric Vehicle Type'])['Electric Vehicle Type'].count().unstack()
df_ev_type
```

```python
df_ev_type.plot()
```

In order to check for outliers, I decided to plot the Electric Range (in Miles) and the Model Year in a box plot. This would allow us to see if there are any outliers that may require further investigation and it also gives us a visual view of the min, max, average, and 75th and 25th percentile.

```python
df_elec[['Electric Range (in Miles)']].boxplot()
```

From the above box plot, we can see that the Electric Range (in Miles) doesn't seem to have any outliers. 

I decided to also make a bar chart about make and count of car for the top fifteen most popular cars using the following code:

```python
df_bar2 = df.groupby(['Make', 'Electric Vehicle Type']).size().unstack().sort_values(by=['Battery Electric (BEV)'], ascending=False)
df_bar2.head(15).plot(kind='bar')
```

In the above chart, we can see that some cars are more popular as Plug In Hybrid ones, compared to Battery Electric ones. We can also see that Nissan and Chevrolet are in the top three in terms of how many electric cars there are. 

Next, we can filter the table by certain attributes of the variables stored. In the code below, I'm filtering by only the object type, which shows a dataframe with columns that are only strings.

```python
df.select_dtypes(include='object')
```
