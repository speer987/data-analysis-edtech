## Data Loading

```python
df_gas = pd.read_excel("../data/raw/Gas_Prices.xls", sheet_name="Data 1", header=2)
df_EVs = pd.read_csv("../data/raw/Electric_Vehicle_Population_Data.csv")
```

## Data Cleaning

#### Gas Prices
We can start by cleaning the gas prices dataset, associated with df_gas.

```python
df_gas.head()
```

We can assign the third row (Date, Washington All Grades Conventional Retail Gasoline Prices) as the new header. We can do this using the following, as the index of the third row is 2. Which is why we write header=2 as one of the parameters in the read_excel function call below.

```python
df_gas.head()
```

I also decided to make the second column's header name of "Washington All Grades Conventional Retail Gasoline Prices (Dollars per Gallon)" to "WA Retail Gasoline Prices (Dollars Per Gallon)" using the following code.

```python
df_gas.rename(columns={'Washington All Grades Conventional Retail Gasoline Prices (Dollars per Gallon)': 'WA Retail Gas Prices (Dollars per Gallon)'}, inplace=True)
df_gas.head()
```

Now, if we'd like to get rid of the 06-30 part in all of the dates, we need to check to make sure that the year is the only value that changes. In order to check that, we can take a look at the column of dates:

```python
df_gas['Date']
```

We can see that the column type is datetime as shown at the bottom of the above output. Additionally, the only values that change are the years, so we can effectively remove the month and days. We can use the following code to extract just the year and set that as the new column's values:

```python
df_gas['Date'] = df_gas['Date'].dt.year
df_gas.head()
```

Although the year can be used as an index because each year is unique, I will leave it as it is because we will be using the year column to merge the two tables later. 

I'll also check to make sure that the data types are right for this table.

```python
df_gas.info()
```

## Electric Vehicle Population Dataset Cleaning

We can also implement the changes I mentioned earlier in the EV Population Dataset.

```python
df_EVs.head()
```

First, we can drop the duplicate rows if there are any in this dataset.

```python
og_len = df_EVs.shape[0]
df_EVs.drop_duplicates()
new_len = df_EVs.shape[0]
print("Original Length:", og_len)
print("New Length:", new_len)
```

Turns out that there weren't any duplicates in the dataset, which is why the length of the dataset didn't change after dropping duplicates.

Now we can check if each VIN number and DOL Vehicle ID is unique.

```python
print("Unique VIN Number Count:", df_EVs['VIN (1-10)'].nunique())
print("Unique DOL Vehicle ID Count:", df_EVs['DOL Vehicle ID'].nunique())
print("Total Original Table Length:", df_EVs.shape[0])
```

Since the number of rows of VIN numbers are not 279780, which was the length of the original dataset, we can drop this column as it cannot be used as an index for the dataframe/table, as this would require that there are a total of 279780 unique VIN numbers.

On the other hand, the DOL Vehicle ID column can be reassigned as the index column, as it seems that there are 279780 unique values, which is the same exact length as the original dataset. We can do this like so:

```python
df_EVs.set_index('DOL Vehicle ID', inplace=True)
df_EVs.head()
```

Now, we can proceed to drop the columns that are unnecessary for this exploration.
- Legislative District
- VIN (1-10)
- 2020 Census Tract

```python
df_EVs.drop(columns=['VIN (1-10)', 'Legislative District', '2020 Census Tract'], inplace=True)
df_EVs.head()
```

Next up, we can change the data saved in the Electric Vehicle Type column to just Plug-in Hybrid, Battery Electric, or None, but we first have to check the column values.

```python
df_EVs.value_counts('Electric Vehicle Type')
```

Now that we've confirmed that there's only two types of values and we don't have to deal with None values, we can replace them like so:

```python
df_EVs['Electric Vehicle Type'].replace({'Plug-in Hybrid Electric Vehicle (PHEV)': 'Plug-in Hybrid (PHEV)', 'Battery Electric Vehicle (BEV)': 'Battery Electric (BEV)'}, inplace=True)
df_EVs.head()
```

We can also change the next column, which is Clean Alternative Fuel Vehicle (CAFV) Eligibility like so:
- Clean Alternative Fuel Vehicle Eligible &rarr; True
- Eligibility Unknown... &rarr; None
- Not Eligible &rarr; False

First, let's check the values of the column:

```python
df_EVs.value_counts('Clean Alternative Fuel Vehicle (CAFV) Eligibility')
```

Since there are only three outcomes, and the column name could be phrased like a question ("Is this car CAFV eligible?"), it makes this column a good candidate to have boolean (True/False) values. We can make this column have boolean values like so:

```python
df_EVs['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].replace({
'Eligibility unknown as battery range has not been researched': None,
'Clean Alternative Fuel Vehicle Eligible': True,
'Not eligible due to low battery range': False,
}, inplace=True)
df_EVs.head()
```

While we're at it, let's also change the column name so it is indeed phrased like a question that can be answered by True/False or None.

```python
df_EVs.rename(columns={'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'Clean Alternative Fuel Vehicle (CAFV) Eligible?'}, inplace=True)
df_EVs.head()
```

In the table now, we can see that the cells that say "Eligibility unknown as battery range has not been researched" in the "CAFV Eligible?" column usually have 0.0 as a value in the next column called "Electric Range". We want to prevent averages getting dragged down for the "Electric Range column", so we'd need to replace these 0.0s with np.nan.

Since there are numeric values, we need to use np.nan instead of 0.0.

```python
df_EVs['Electric Range'] = df_EVs['Electric Range'].replace(0.0, np.nan)
df_EVs.head()
```

We can also change the postal code to integers, because they are normally not decimal numbers (or floats in this case).

Electric Ranges can also be changed to integers, as there doesn't seem to be numbers after the decimal point for each Electric Range. But first, let's check if we need nullable integers like so:

```python
df_EVs.info()
```

From the above output, we can see that both the "Postal Code" and "Electric Range" column has a couple of null cells, so we would have to cast these columns to "Int64", which makes null values into \<NA>. If all values were filled (none of them being Null or None), we could just cast to int64.

```python
df_EVs['Electric Range'] = df_EVs['Electric Range'].astype('Int64')
df_EVs['Postal Code'] = df_EVs['Postal Code'].astype('Int64')
df_EVs.head()
```

We also need to change the column name of "Electric Range" to "Electric Range (in Miles)". 

```python
df_EVs.rename(columns={'Electric Range': 'Electric Range (in Miles)'}, inplace=True)
df_EVs.head()
```

Now we need to inspect if it is worth changing cells of the Electric Utility (like how "CITY OF SEATTLE - (WA)|CITY OF TACOMA - (WA)" can be changed to "CITY OF SEATTLE/TACOMA"). 

In order to check if this is possible, let's see if there are other values in the column that are long like this. 

```python
df_EVs['Electric Utility'].unique()
```

From the above output, we can see that there are quite a few long ones... For this analysis, I will leave this column as is, as it may require more complexity in my code, and may change the output of calculations done on other rows.

Finally, let's check to make sure that the data types for each column are right before proceeding to merge the two tables.

```python
df_EVs.info()
```

## Merging The Two Datasets

We can finally merge the two datasets since we have finished cleaning the data!

Now that we've cleaned the datasets, we can merge/join them based on Year column in the df_gas and the Model Year in df_EVs. Both of these columns are of the "int64" datatype, so merging should work. We can do this using the following:

```python
df_merged = df_EVs.merge(df_gas, how="left", left_on='Model Year', right_on='Date', validate=None)
df_merged.head()
```

Since there are now two columns for the date after merging, we can drop one of the columns.

```python
df_merged.drop(columns=['Date'], inplace=True)
df_merged.head()
```
