### Data Cleaning

#### Gas Prices
We can start by cleaning the gas prices dataset, associated with df_gas.

```python
df_gas.head()
```

We can assign the third row (including the header) as the new header. We can do this using the following, as the index of the third row is 2. Which is why we write header=2 as one of the parameters in the read_excel function call below.

```python
df_gas = pd.read_excel("Gas_Prices.xls", sheet_name="Data 1", header=2)
df_gas.head()
```

I also decided to make the second column's header name of "Washington All Grades Conventional Retail Gasoline Prices (Dollars per Gallon)" to "WA Retail Gasoline Prices (Dollars Per Gallon)" using the following code.

```python
df_gas.rename(columns={'Washington All Grades Conventional Retail Gasoline Prices (Dollars per Gallon)': 'WA Retail Gas Prices (Dollars per Gallon)'}, inplace=True)
df_gas.head()
```

Now, if we'd like to get rid of the 06-30 part in all of the dates, we need to check to make sure that the year is the only value that changes. In order to check that, we can take a look at the column of dates:

```python
df_gas['Date']
```

We can see that the column type is datetime as shown at the bottom of the above output. Additionally, the only values that change are the years, so we can effectively remove the month and days. We can use the following code to extract just the year:

```python
df_gas['Date'] = df_gas['Date'].dt.year
df_gas.head()
```

Although the year can be used as an index, I will leave it alone because we will be using the year column to merge the two tables later.

#### Electric Vehicle Population Dataset Cleaning

We can also implement the changes I mentioned earlier in the EV Population Dataset.

```python
df_EVs.head()
```

First, we can drop the duplicate rows if there are any in this dataset.

```python
og_len = df_EVs.shape[0]
df_EVs.drop_duplicates()
new_len = df_EVs.shape[0]
print("Original Length:", og_len)
print("New Length:", new_len)
```

Turns out that there weren't any duplicates in the dataset, which is why the length of the dataset didn't change after dropping duplicates.

Now we can check if each VIN number and DOL Vehicle ID is unique.

```python
VIN = df_EVs['VIN (1-10)']
VIN.nunique()
VIN.value_counts()
```

```python
df_EVs['DOL Vehicle ID'].nunique()
```

Turns out that the VIN (1-10) column cannot be used as the id of this row since the unique entries are only about 17000 entries, which is much lower than the total of 279780 entries. So, we can drop the VIN (1-10) column, since it doesn't seem to be relevant to this exploration.

On the other hand, the DOL Vehicle ID column can be reassigned as the index column, as it seems that there are 279780 unique values, which is the same exact length as the original dataset. We can do this like so:

```python
df_EVs.set_index('DOL Vehicle ID', inplace=True)
df_EVs.head()
```

Now, we can proceed to drop the columns that are unnecessary for this exploration.
- Legislative District
- VIN (1-10)
- 2020 Census Tract

```python
df_EVs.drop(columns=['VIN (1-10)', 'Legislative District', '2020 Census Tract'], inplace=True)
df_EVs.head()
```

Next up, we can change the data saved in the Electric Vehicle Type column to just Plug-in Hybrid, Battery Electric, or NaN, but we first have to check the column values.

```python
df_EVs.value_counts('Electric Vehicle Type')
```

Now that we've confirmed that there's only two types of values, we can replace them like so:

```python
df_EVs['Electric Vehicle Type'].replace({'Plug-in Hybrid Electric Vehicle (PHEV)': 'Plug-in Hybrid (PHEV)', 'Battery Electric Vehicle (BEV)': 'Battery Electric (BEV)'}, inplace=True)
df_EVs.head()
```

We can also change the next column, which is Clean Alternative Fuel Vehicle (CAFV) Eligibility like so:
- Clean Alternative Fuel Vehicle Eligible -> True
- Eligibility Unknown... -> NaN or NULL
- Not Eligible -> False

First, let's check the values of the column:

```python
df_EVs.value_counts('Clean Alternative Fuel Vehicle (CAFV) Eligibility')
```

Now, we can replace those values accordingly:

```python
df_EVs['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].replace({
'Eligibility unknown as battery range has not been researched': None,
'Clean Alternative Fuel Vehicle Eligible': True,
'Not eligible due to low battery range': False,
}, inplace=True)
df_EVs.head()
```

A couple of code blocks ago, we also saw that the CAFV Eligibility column had some data that said: "Eligibility unknown as battery range has not been researched". For these columns you can see that the battery range is noted at 0.0 in the next column for electric range. We want to prevent averages getting dragged down, so we'd need to replace these 0.0s with np.nan.

I noticed when I used None the first time, the 0.0s were getting replaced with the values from the previous rows from above, which made the data have the wrong values. Upon further research, I learned that whenever you deal with numeric values, you need to use np.nan.

```python
df_EVs['Electric Range'] = df_EVs['Electric Range'].replace(0.0, np.nan)
df_EVs.head()
```

Next, we can change the postal code, and possibly the electric range to integers. In the case of electric ranges, we need to check if it has any numbers after the decimals to make sure that we won't lose any numerical precision.

```python
# I got stuck at this point, so I had to use an LLM to check for rows where there are numbers with a decimal point and not null:
mask = (~np.isclose(df_EVs["Electric Range"] % 1, 0)) & pd.notna(df_EVs["Electric Range"])
df_EVs[mask]
```

Using the above output, we can see that there are no rows where there is a number after the decimal point. Therefore, we can use the following code to remove the zeroes from the end of each number:

```python
df_EVs['Electric Range'] = df_EVs['Electric Range'].astype('Int64')
df_EVs.head()
```

As a general principle, postal codes don't have decimal points, so we can remove the 0 from the end by casting each postal code, which is a float, to an int. We can check to see if we need to use 'Int64', based on how many null values there are in the dataset using the .info() function.

```python
df_EVs.info()
```

The above information shows us that there are a couple of entries that are null in the postal code as well, which means we ought to use 'Int64' casting.

```python
df_EVs['Postal Code'] = df_EVs['Postal Code'].astype('Int64')
df_EVs.head()
```

Now that we've cleaned the datasets, we can merge/join them based on Year column in the df_gas and the Model Year in df_EVs. We can do this using the following:

```python
df_gas.info()
```

```python
print(df_EVs.info())
print(df_gas.info())

df_merged = pd.merge(df_EVs, df_gas, how='left', on='Date', validate=None)
df_merged.head()
```
